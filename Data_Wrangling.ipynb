### Dealing with Missing Values
* When no data values is stored for a feature for a particular observation, it is known as a `missing value`.
* Missing values are represented as `NaN`, `None`, `?`, `N/A`, `0` or just a `Blank Cell`.

#### How to deal with Missing values?
There are typical options we can consider:
1. The first is to check if the person or group that collected the data can go back and find what the actual value should be.
2. Another possibility is just to remove the data where the value is missing. This is known as `listwise deletion` or `case deletion`.

#### Drop the missing values.
* Drop the variables
* Drop the data entry

If we don't have a lot of observation with missing data, usually dropping the particular entry is the best.

```python
# To drop missing values in python we use dropna() function. 
# By default, it drops rows with missing values.
# To drop columns, we need to provide axis=1 as an argument to the dropna() function.
df.dropna(subset=["price"], axis=0, inplace=True)
# Setting the argument inplace=True allows us to update the dataset df.
```

#### Replace the missing values.
* If we are removing the data entry, we are losing the information for the other variables in that entry.
* Replacing data is the better, since no data is wasted.
* It is less accurate since we need to make assumptions about the data.
Replace the missing values with:
1. Mean or average
2. Frequency
3. Other functions

```python
# To replace missing values in python we use replace() function.
# dataFrame.replace(missing_value, new_value)
# missing_value: the value we want to replace
# new_value: the new value we want to replace with
# We can use mean, median or mode to replace missing values.
# We can also use .mean(), .median() or .mode() functions to calculate the mean, median or mode of the column.
import numpy as np
df.replace('?', np.nan, inplace=True)
```

# Data Notmalization in Python
Sometimes during analysis due to large difference in the ranges of two columns in a dataset, the higher range columns intrinsically influence the result more, due to its larger value, but this doesn't necessarily mean that it is more important as a predictor. So we normalize the data to bring all the columns to a more similar range.

* There are several ways to normalize data:
    * Simple Feature Scaling: Just divide each value by the maximum value for that feature.
    * Min-Max: Takes each value, x-old, substracted from the minimum value of that feature, and then divide by the range of that feature.
    * Z-Score: In this formula, for each calue, we sustract Mu which is the average of all the values, and then divide by the standard deviation (Sigma) of all the values.
    * Decimal Scaling
    * Softmax

# Binning in Python
* Binning: Grouping of values into "bins.
* Converts numeric into categorical variables.
* Groups a set of numerical values into a set of "bins".
* Binning is good for:
    * Comparing groups of data
    * Identifying trends
    * Identifying outliers
* Binning can be used to:
    * Smoothing
    * Aggregating
    * Visualizing
* Binning Methods:
    * Equal Width Binning
    * Equal Frequency Binning
* Something, binning can improve accuracy of the predictive models.
* Something, we use data binning to group a set of numerical values into a set of bins to have a better understanding of the data distribution.
* In Python, we can easily implement binning: 
    * We would like 3 bins of equal binwidth, so we need 4 numbers as dividers that are equal distance apart.
    * First, we use the numpy function `linspace` to return the array "bins" that contains 4 equally spaced numbers over the specified interval of the price.
    * We create a `list` named `group_names` that contains the different bin names.
    * We use pandas function `cut` to segment and sort the data values into bins.

```python
bins = np.linspace(min(df["price"]), max(df["price"]), 4)
group_names = ['Low', 'Medium', 'High']
df['price-binned'] = pd.cut(df['price'], bins, labels=group_names, include_lowest=True)
```
